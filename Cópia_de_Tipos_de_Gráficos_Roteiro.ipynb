# def GraficoLinha(style)

```python
def GraficoLinha(style = 'ggplot'):
  #estilo de cor
  plt.style.use(style) 
  # Aumentar o tamanho da fonte
  plt.rcParams.update({'font.size': 16})

  # Criar um gráfico de linha
  plt.figure(figsize=(16, 12))

  #descompactar lista
  produto1, = plt.plot(df['Meses'], df['Produto 1'], label='Produto 1', marker='o') 
  produto2, = plt.plot(df['Meses'], df['Produto 2'], label='Produto 2', marker='o')
  produto3, = plt.plot(df['Meses'], df['Produto 3'], label='Produto 3', marker='o')


  plt.legend([produto1, produto2, produto3], ['Produto 1', 'Produto 2', 'Produto 3'])
  plt.title('Vendas de Produtos (2023)')
  plt.xlabel('Mês')
  plt.ylabel('Vendas')

  plt.savefig('GraficoLinha.svg')
  # Mostrar o gráfico
  plt.show()  
```

# Gráfico de Caixa (Box Plot)

```python
# Dados
distribuicao = np.random.normal(40, 20, 120)

# Criar o gráfico de caixa
plt.boxplot(distribuicao)

plt.title('Distribuição Personalizada')
plt.xlabel('Dados')
plt.ylabel('Valores')

# Mostrar o gráfico
plt.show()
```

# Explique o Gráfico acima -- Pesquise
R: Esse grafico geral que mostra as medidas de quartis, mediana, amplitude. No boxplot a parte central do gráfico contém os valores que estão entre o primeiro quartil e o terceiro quartil.

```python
!gdown 1GvcLp9qKfSI2o0CHvXjsGXDRTZ91SGy9
```

```python
df = pd.read_csv('Pontuacao.csv')
df
```

```python
# Criar o gráfico de violino
equipes = df.columns.tolist()
plt.figure(figsize=(8, 6))
plt.violinplot(df)

# Personalizar o gráfico
plt.title('Distribuição da Pontuação por Equipe')
plt.xlabel('Equipes')
plt.ylabel('Pontuação')

plt.xticks(np.arange(1, len(equipes)+1), equipes)

# Mostrar o gráfico
plt.show()
```

# Explique o gráfico acima, qual é a diferença da Equipe C da Equipe D.
R:

Cada gráfico representa a distribuição da pontuação da equipe.
Quanto mais largo o violino, maior é a frequência de pontuações em uma determinada faixa.
A linha branca no centro de cada violino representa a mediana da distribuição da pontuação para cada equipe.
As extremidades dos violinos são chamadas de "arestas". Elas mostram a variação das pontuações em cada equipe. Quanto mais longas as arestas, maior é a dispersão dos valores.
Os pontos individuais acima e abaixo dos violinos representam valores discrepantes (outliers) que estão fora do intervalo da maioria das pontuações.
